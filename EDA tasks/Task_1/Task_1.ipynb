```python
def unique_values(col):
    print('unique values: {}\n'.format(df[col].unique()))
```

```python
def salary_statistics_by_feature(col):
    with pd.option_context('display.colheader_justify','left'):
        print('{}'.format(df.groupby(col)['Salary'].describe()))
```

```python
def NONE_values_by_feature(col):
    print('No. of NONE values: {}'.format(sum(n=='NONE' for n in df[col].values)))
```

```python
data_features = ['Designation', 'JobCity', 'Gender', 'Degree', 
                  '10percentage', '12percentage', 'CollegeTier', 'Salary']
```

```python
for col in data_features:
    print('\n\n\n\033[1m{}:\033[0m'.format(col))
    NONE_values_by_feature(col)
    unique_values(col)
    salary_statistics_by_feature(col)
```
